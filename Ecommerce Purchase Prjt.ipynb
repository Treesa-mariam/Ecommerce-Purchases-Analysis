```python
#Dataset - https://www.kaggle.com/datasets/utkarsharya/ecommerce-purchases
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Ecommerce Purchases')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram for purchase prices
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Purchase Price'], kde=True)
plt.title("Distribution of Purchase Prices")
plt.xlabel("Purchase Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for language distribution
language_counts = df['Language'].value_counts()
language_counts.plot(kind='bar', figsize=(8, 5), color='skyblue')
plt.title("Language Distribution")
plt.xlabel("Language")
plt.ylabel("Number of Customers")
plt.show()
```

```python
#Segmentation Analysis
```

```python
# Creating segments based on purchase price
def segment_price(price):
    if price < 50:
        return 'Low'
    elif 50 <= price < 100:
        return 'Medium'
    else:
        return 'High'

df['Price Segment'] = df['Purchase Price'].apply(segment_price)
```

```python
# Count of each segment
segment_counts = df['Price Segment'].value_counts()
print("Customer Segments:\n", segment_counts)
```

```python
#Purchase Time Analysis
```

```python
df.columns
```

```python
# Count of purchases in AM vs PM
am_pm_counts = df['AM or PM'].value_counts()
print("Purchases by AM/PM:\n", am_pm_counts)

# Plotting the AM vs PM purchases
plt.figure(figsize=(8, 5))
am_pm_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title("Purchases in AM vs PM")
plt.xlabel("Time of Day")
plt.ylabel("Number of Purchases")
plt.show()

```
